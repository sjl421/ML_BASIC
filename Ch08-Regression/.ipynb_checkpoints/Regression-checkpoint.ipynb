求解回归方程中的回归系数（即权重）的过程称为回归

## 1. 标准线性回归

**预测函数**：也称为回归方程：$\hat{y} = X*w$ 其中 $w$ 称为回归系数（或权重）。

**代价函数**：$J(w)=\frac{1}{m}\sum_{i=1}^{m}(\hat{y}-y)^2$

**求解权重的方法**：
1. 正规方程：通过 $\frac{\partial J(w)}{\partial w}=0$ 得到 $w = (X^TX)^{-1}X^Ty$
2. 使用最优化算法，如梯度下降

````python
import numpy as np
import matplotlib.pyplot as plt

def load_data(filename):
    fr = open(filename)
    lines = fr.readlines()
    m = len(lines)
    n = len(lines[0].strip().split('\t'))
    dataset = np.zeros((m,n))
    for i in range(m):
        l = lines[i]
        dataset[i,:] = l.strip().split('\t')
    return dataset

dataset = load_data('./dataset/ex0.txt')
x = dataset[:,:-1]
y = dataset[:,-1].reshape((x.shape[0],1))
w = np.dot(np.linalg.inv(np.dot(x.T, x)), np.dot(x.T, y))
print(w)

plt.scatter(x[:,1].flatten(), y.flatten(), 5)
x = np.array([
    [1., 0.],
    [1, 0.5],
    [1, 1.0]
])
y_hat = np.dot(x, w)
plt.plot(x[:,1].flatten(), y_hat.flatten(), c='red')
plt.show()
````

## 2. 局部加权线性回归（LWLR）

**为什么使用局部加权线性回归**：（TODO）能模拟出非线性

**LWLR的核心思想** ：给待预测点附近的每个点赋予一定的权重。回归系数 $w = (X^TWX)^-1X^TWy$ 其中 $W$ 是一个矩阵，用来给每个样本点赋予权重。

**样本权重矩阵**$W$ : 是一个 (m,m) 的矩阵，且只有对角元素非零，其余元素全为零。对角元素 $W(i,i)$ 表示当前待预测点与第 i 个样本点的权重。当前待预测点 $x$ 与第 i 个样本点越近， $W(i,i)$ 越大。

**计算样本权重**$W$ **公式** ：计算样本权重的算法称为 LWLR 的核。选择不同核，计算方式就不同。常见的核是高斯核。 $W(i,i) = exp(\frac{|x^{(i)}-x|^2}{-2k^2})$ 其中 $k$ 决定了对附近的点赋予多大的权重，是由用户设置的参数。

**优点**：
1. 能拟合非线性

**缺点**：
1. 计算成本高；由于样本权重矩阵对于每个待预测点都不一样，那么，对于每个待预测点的回归系数都不一样。意味着，对每个待预测点都需要重新样本权重矩阵和回归系数。 *(详见下面代码，尤其是第 35-36 行)*
2. 需要保留所有训练数据；标准的线性回归中用训练集得到回归系数 $w$ 后，就可以直接运用进行预测。而 LWLR 必须保留所有训练数据，因为要求解样本权重矩阵。 

````python
def train_lwlr(test_vector, x, y, k=1.0):
    '''
    训练基于高斯核的局部甲醛线性回归
    :param test_vector: [np.array(1,n)] 待预测点
    :param x: [np.array(m,n)] 特征矩阵
    :param y: [np.array(m,1)] 标签向量
    :param k: [float] 计算样本权重矩阵的参数
    :return : [np.array(n,1)] w 回归系数
    '''
    m = x.shape[0]
    assert(y.shape == (m,1))
    sample_weights = np.eye((m))  # 样本矩阵
    # 基于高斯核计算样本矩阵
    for j in range(m):
        diff_mat = test_vector - x[j,:].reshape(test_vector.shape)  # 分子
        sample_weights[j,j] = np.exp(np.dot(diff_mat, diff_mat.T) / (-2.0*k*k))

    # 计算 w 
    xTx = np.dot(np.dot(x.T, sample_weights), x)
    w = np.dot(np.dot(np.dot(np.linalg.inv(xTx), x.T), sample_weights), y)
    return w

def test_lwlr(test_x, x, y, k=1.0):
    '''
    使用基于高斯核的局部甲醛线性回归进行预测
    :param test_x: [np.array(m',n)] 测试特征矩阵
    :param x: [np.array(m,n)] 特征矩阵
    :param y: [np.array(m,1)] 标签向量
    :param k: [float] 计算样本权重矩阵的参数
    :return : [np.array(m,1)] 预测结果
    '''
    m,n = x.shape
    y = y.reshape((m,1))
    y_hat = np.zeros(y.shape)
    for i in range(test_x.shape[0]):
        w = train_lwlr(test_x[i,:].reshape((1,n)), x, y, k)
        y_hat[i,:] = np.dot(test_x[i,:], w)
    return y_hat
        
# 1. 读取数据过程
dataset = load_data('./dataset/ex0.txt')
x = dataset[:,:-1]
y = dataset[:,-1].reshape((x.shape[0],1))
# 2. 训练lwlr并得到预测结果
y_hat = test_lwlr(x, x,y,0.003)
# 3. 可视化结果
plt.scatter(x[:,1].flatten(), y.flatten(), 5)
# 对 x 按值排序
srtind = x[:,1].argsort(0)
xsort = np.zeros(x.shape)
ysort = np.zeros(y_hat.shape)
index = 0
for i in srtind:
    xsort[index, :] = x[i, :]
    ysort[index, :] = y_hat[i, :]
    index += 1
plt.plot(xsort[:,1].flatten(), ysort.flatten(), c='red')
plt.show()
````

````python
# 示例：预测鲍鱼的年龄
# 1. 读取数据过程
dataset = load_data('./dataset/abalone.txt')
x = dataset[:,:-1]
y = dataset[:,-1].reshape((x.shape[0],1))
# print(x.shape, y.shape)

y_hat01 = test_lwlr(x[0:99,:], x[0:99,:], y[0:99,:], 0.1)
print('k=0.1 the train error is ', np.sum((y_hat01-y[0:99,:])**2) / x.shape[0])
y_hat01 = test_lwlr(x[100:199,:], x[0:99,:], y[0:99,:], 0.1)
print('k=0.1 the test  error is ', np.sum((y_hat01-y[100:199,:])**2) / x.shape[0])
print('================')
y_hat1 = test_lwlr(x[0:99,:], x[0:99,:], y[0:99,:], 1.0)
print('k=1.0 the train error is ', np.sum((y_hat1-y[0:99,:])**2) / x.shape[0])
y_hat01 = test_lwlr(x[100:199,:], x[0:99,:], y[0:99,:], 1.0)
print('k=1.0 the test  error is ', np.sum((y_hat01-y[100:199,:])**2) / x.shape[0])
print('================')
y_hat10 = test_lwlr(x[0:99,:], x[0:99,:], y[0:99,:], 10)
print('k=10. the train error is ', np.sum((y_hat10-y[0:99,:])**2) / x.shape[0])
y_hat01 = test_lwlr(x[100:199,:], x[0:99,:], y[0:99,:], 10)
print('k=10. the test  error is ', np.sum((y_hat01-y[100:199,:])**2) / x.shape[0])
````

明显发现，核的 $k$ 越小将得到更低的训练误差，更高的测试误差。即 **k 越小，越容易发现过拟合**

## 3. 缩减系数来“理解”数据：岭回归、lasso 以及前向逐步回归

当 $n > m$ 时，此时特征矩阵 $X$ 不可逆。之前的正规方程就不能使用了。如何解决？ 解决方案：

1. 降维（不是本实验采用的方法）
2. 缩减回归系数，通过一定的约束

上述两种方法都会或多或少造成：减小了方差(varience)、增加了偏差(bias)

### 3.1 岭回归

**岭回归中的正规方程** ：$w = (X^TX + \lambda I)^{-1}X^Ty$ ,即在原先的正规方程增加了 (n,n) 的单位矩阵。其中 $\lambda$ 时用户定义的参数。

$\lambda$ **的意义** ：引入 $\lambda$ 来限制所有 $w_i$ 之和，通过引入该惩罚项，能够减少不重要的参数，这个技术在统计学中也叫做缩减。

**理解岭的意思** : 跟单位矩阵有关。单位矩阵中除了对角元素为 1 ，其他元素为 0 。对角元素构成一个“岭”。*（无语了）*

````python
def ridge_regress(x, y, lamb=0.2):
    '''
    训练岭回归
    :param x: [np.array(m,n)] 特征矩阵
    :param y: [np.array(m,1)] 标签向量
    :param lamb: [float] 系数
    :return : [np.array(1,n)] w 回归系数
    '''
    I = np.eye((x.shape[1]))
    temp = np.dot(x.T, x)
    temp = temp + lamb*I
    temp = np.linalg.inv(temp)
    w = np.dot(np.dot(temp, x.T), y)
    return w.reshape((1, x.shape[1]))

# 1. 读取数据过程
dataset = load_data('./dataset/abalone.txt')
x = dataset[:,:-1]
y = dataset[:,-1].reshape((x.shape[0],1))

# 2. 标准化数据
y = y - np.mean(y, 0)
x = (x - np.mean(x, 0)) / np.var(x, 0)

# 3. 训练,尝试不同的 lambda
w_matrix = np.zeros((30, x.shape[1]))
for i in range(30):  # 尝试不同的 lambda
    w = ridge_regress(x, y, np.exp(i-10))
    w_matrix[i,:] = w

# 4. 可视化结果
# print(w_matrix)
import matplotlib.pyplot as plt
plt.plot(w_matrix)  # (30,8) 回归系数向量中有8个回归系数，尝试了30个lambda，
plt.ylabel('w_i')
plt.xlabel('i')
plt.show()
    
````

解释 ：上图中，每一条线代表一个回归系数 $w_i$ 随着不同的 $\lambda$ 值的变化趋势。从左往右的方面是 $\lambda$ 增大的方向。当 $\lambda$ 增大到一定大小时，所有回归系数缩减成 0 。为了定量地找到最佳参数值，还需要进行交叉验证。

### 3.2 lasso

配合着最优化算法（如梯度下降）使用，在最小化代价函数 $min~J(w) = min~\frac{1}{m} \sum_{i=1}^{m} (y_i-\hat{y_i})^2$ 时，增加一个约束 $\sum_{j=1}^{n}w_j^2 \leq \lambda$ 或者 $\sum_{j=1}^{n}|w_j| \leq \lambda$ 其中，$\lambda$ 是用户设定的参数。当 $\lambda$ 足够小的时候，一些系数会因此被迫缩减到 0.


### 3.3 前向逐步回归

它属于一种贪心算法，即每一步都尽可能减少为误差。

算法的伪代码：

```
数据标准化，使其分布满足满足 0 均值核单位方差
对每轮迭代过程中：
    设置当前最小误差 min_error 为正无穷
    对每个特征：
        增大或缩小该特征：
            改变一个系数更新一个回归系数向量 w
            计算新的 w 下的误差
            如果此误差小于当前最小误差 min_error ：设置 best_w 等于当前的 w
    将 w 设置为 best_w
```

````python
def train_stagewise(x, y, eps=0.01, iter_num=100):
    '''
    前向逐步法训练
    :param x: [np.array(m,n)] 特征矩阵
    :param y: [np.array(m,1)] 标签
    :param eps: [float] 步长，表示每次迭代需要调整的步长
    :param iter_num: [int] 迭代次数
    :return : [np.array(iter_num, n)] w_matrix 每次迭代得到的 w 所组成的矩阵
    '''
    m,n = x.shape
    assert(y.shape == (m,1))
    # 1. 数据标准化
    y = y - np.mean(y,0)
    x = (x - np.mean(x,0)) / np.var(x,0)
    w_matrix = np.zeros((iter_num, n))
    w = np.zeros((n,1))
    best_w = w.copy()
    for i in range(iter_num):
#         print(w.T)
        min_error = float('inf')
        for j in range(n):
            for sign in [-1, 1]:
                cur_w = w.copy()  # 当前的 w
                cur_w[j,:] += eps*sign
                cur_y = np.dot(x, cur_w)  # 当前的 w 下产生的预测结果
                cur_error = np.sum((cur_y-y)**2) / m  # 计算当前误差
                if cur_error < min_error:
                    min_error = cur_error
                    best_w = cur_w
        w = best_w.copy()
        w_matrix[i,:] = best_w.T
    
    return w_matrix

# 1. 读取数据过程
dataset = load_data('./dataset/abalone.txt')
x = dataset[:,:-1]
y = dataset[:,-1].reshape((x.shape[0],1))
w_matrix = train_stagewise(x,y, eps=0.005, iter_num=1000)
import matplotlib.pyplot as plt
plt.plot(w_matrix)
plt.xlabel('iter num')
plt.ylabel('w_i')
plt.show()
````
