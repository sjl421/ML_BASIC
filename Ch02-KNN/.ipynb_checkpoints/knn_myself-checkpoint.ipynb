## KNN 综述

简单地说，KNN 算法采用测量不同特征值之间的距离方法进行分类。

属于监督学习，分类算法。

**思想：计算新输入样本与样本数据集中每个样本的距离，只选择样本数据集中前 k 个距离最小的（即最相似的）数据。最后，这 k 个最相似数据中出现次数最多的分类，作为新样本的分类。**

## KNN 核心算法

```python
import numpy as np
import matplotlib.pyplot as plt
import operator,os

# 解决Matplotlib输出中文显示问题
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
def KNN_classify(inX, feature_matrix, labels, k):
    '''
    kNN 分类算法
    采用欧氏距离计算样本间的距离
    :param inX: 输入样本特征向量(1,n)
    :param feature_matrix: 数据集的特征矩阵(m,n)
    :param labels: 数据集对应的标记向量(m,)
    :param k: 实数
    :return：输入样本的预测类别
    '''
    m = feature_matrix.shape[0]  # 数据集大小
    n = feature_matrix.shape[1]  # 特征数
    assert(inX.shape == (1, n))
    assert(labels.shape == (m,))
    inX_matrix = np.tile(inX, (m,1))  # 将新样本从向量扩展为(m,n)矩阵
    
    # 计算新样本与数据集的欧氏距离
    distance_matrix = inX_matrix - feature_matrix
    distance_matrix = distance_matrix ** 2
    distance_vector = np.sum(distance_matrix, axis=1) ** 0.5
    
    # 选择距离最小的 k 个点,并统计这 k 个最相似数据中类别出现次数
    sorted_dist_indexs = distance_vector.argsort()
    class_counts = {}
    for i in range(k):
        label = labels[sorted_dist_indexs[i]]
        class_counts[label] = class_counts.get(label, 0) + 1
#     print(class_counts)
    # 找这 k 个最相似数据中出现次数最多的分类
    rows_by_count = sorted(class_counts.items(), key=lambda c:c[1], reverse=True)  # 字典按次数大小排序
    return rows_by_count[0][0]
    
```

上面这个分类算法就是 KNN 的核心。如何运用 KNN 算法就是将数据整理好，输入到这个分类算法中就可以使用。

下面展示 KNN 使用示例。

```python
# 创造假数据来测试 KNN_classify()
def create_dataset():
    group = np.array([
        [1.0, 1.1],
        [1.0, 1.0],
        [0, 0],
        [0, 0.1]
    ])
    labels = ['A', 'A', 'B', 'B']
    return group,labels

group, labels = create_dataset()
print('the real label of [0,0] is B and knn predict label is ', KNN_classify(np.array([[0,0]]), group, np.array(labels), 3))
```

## KNN 算法实战

### 1. 简单的实战：约会网站预测

用书中的数据集跑 KNN 算法。首先是观察数据，然后对数据预处理

```python
def file2matrix(filename):
    '''
    将数据集从文件中读取，并转换为(m,3)矩阵
    :param filename: 完整文件名
    :return：(m,3)特征矩阵和(m,1)的标签向量
    '''
    fr = open(filename)
    all_lines = fr.readlines()
    features_matrix = np.zeros((len(all_lines), 3))
    labels = np.zeros((len(all_lines), 1))
    index = 0
    for line in all_lines:
        line = line.strip().split('\t')
        features_matrix[index, :] = line[0:-1]
        labels[index, :] = line[-1]
        index += 1
    return features_matrix, labels

features_matrix, labels = file2matrix('./dataset/datingTestSet2.txt')

# 可视化数据
plt.subplot(1,2,1)
plt.scatter(features_matrix[:,1], features_matrix[:,2], 15.0*labels.flatten(), 15.0*labels.flatten())
plt.xlabel('玩视频游戏所耗时间百分比')
plt.ylabel('每周消费的冰淇淋公升数')
# plt.show()
plt.subplot(1,2,2)
plt.scatter(features_matrix[:,0], features_matrix[:,1], 15.0*labels.flatten(), 15.0*labels.flatten())
plt.xlabel('每年获取的飞行常客里程数')
plt.ylabel('玩视频游戏所耗时间百分比')
plt.show()
```

```python
def norm_data(dataset):
    '''
    归一化特征
    采用 (value - min) / (max - min).
    '''
    min_vector = dataset.min(0)  # 最小值向量(n,)
    max_vector = dataset.max(0)  # 最大值向量(n,)
    min_matrix = np.tile(min_vector, (dataset.shape[0],1))
    max_matrix = np.tile(max_vector, (dataset.shape[0],1))
    dataset_norm = np.zeros(dataset.shape)
    dataset_norm = (dataset - min_matrix) / (max_matrix-min_matrix)
    return dataset_norm

features_matrix_norm = norm_data(features_matrix)
# 可视化数据
plt.subplot(1,2,1)
plt.scatter(features_matrix_norm[:,1], features_matrix_norm[:,2], 15.0*labels.flatten(), 15.0*labels.flatten())
plt.xlabel('玩视频游戏所耗时间百分比')
plt.ylabel('每周消费的冰淇淋公升数')
# plt.show()
plt.subplot(1,2,2)
plt.scatter(features_matrix_norm[:,0], features_matrix_norm[:,1], 15.0*labels.flatten(), 15.0*labels.flatten())
plt.xlabel('每年获取的飞行常客里程数')
plt.ylabel('玩视频游戏所耗时间百分比')
plt.show()
```

```python
# knn 针对约会网站的完整测试代码
ratio = 0.10  # 测试集占数据集的百分比
features_matrix, labels = file2matrix('./dataset/datingTestSet2.txt')
features_matrix_norm = norm_data(features_matrix)
m = features_matrix_norm.shape[0]  # 数据集大小
error_num = 0  # 错误分类的数量
for i in range(int(m*ratio)):  # 取数据集的前 ratio 的样本作为测试集
    predit = KNN_classify(features_matrix_norm[i,:].reshape(1,3), features_matrix_norm[int(m*ratio):,:], 
                          labels[int(m*ratio):,:].flatten(),3)
#     print('the real label is {} and the predited label is {}'.format(labels[i], predit))
    if predit != labels[i]:
        error_num = error_num + 1
        print('No.{} : the real label is {} and the predited label is {}'.format(i, labels[i], predit))
print('the error rate is  {}'.format(error_num/m))
```

### 2. 复杂点的实战：手写识别系统

输入数据是 32 * 32 的图像。识别数字0-9. 标签信息位于文件名中。

```python
def img2vector(filename):
    '''
    将32*32图像读入数据，并转换为(1,1024)向量。
    :return : (1,1024) 向量
    '''
    vector = np.zeros((1,1024))
    fr = open(filename)
    for i in range(32):
        linestr = fr.readline()
        for j in range(32):
            vector[0,32*i+j] = int(linestr[j])
    return vector

# 准备数据
train_labels = []
train_dir = os.listdir('./dataset/digits/trainingDigits')
train_m = len(train_dir)
train_matrix = np.zeros((train_m, 1024))
for i in range(train_m):
    filename = train_dir[i]
    train_labels.append(int(filename.split('.')[0].split('_')[0]))
    train_matrix[i,:] = img2vector('./dataset/digits/trainingDigits/' + filename)
    
test_labels = []
test_dir = os.listdir('./dataset/digits/testDigits')
test_m = len(test_dir)
test_matrix = np.zeros((test_m, 1024))
for i in range(test_m):
    filename = test_dir[i]
    test_labels.append(int(filename.split('.')[0].split('_')[0]))
    test_matrix[i,:] = img2vector('./dataset/digits/testDigits/' + filename)

# print(train_labels, test_labels)    
    
error_num = 0
index = 0
for i in range(test_m):
    predit = KNN_classify(test_matrix[i,:].reshape(1,1024), train_matrix, np.array(train_labels),3)
#     print('the real label is {} and the predited label is {}'.format(labels[i], predit))
    if predit != test_labels[i]:
        error_num = error_num + 1
        print('No.{} : the real label is {} and the predited label is {}'.format(i, test_labels[i], predit))
print('the error rate is  {}'.format(error_num/test_m))
```

实际上，用 KNN 识别计算成本非常大，效率不高。一个新样本要与 2k 的样本计算距离（因为数据集大小是 2k），每个距离计算包括了 1024 个维度浮点运算（因为一个样本在1024个特征）。由于有900个测试集，那么总的计算成本是 900 * 2000 * 1024 。并且随着数据集和测试样本的不断增加，计算成本也成倍增长。

## 小结

KNN 优点在于

1. 不需要训练模型，直接计算输入样本与数据集的距离即可。
2. 对异常值不敏感。因为你只取距离最近的 k 个数据。

KNN 缺点也很明显：

1. 计算成本高。因为输入样本要与数据集的每个样本计算一次距离
2. 空间复杂度高。因为虽然不需要训练模型，但要完整保存好数据集。
